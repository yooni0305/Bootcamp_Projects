```python
secret_word = "Flower"
display_word =["_"]*len(secret_word)  # ["문자열","문자열"] =list
#print(display_word)
guessed_letter = [] # list
mistake_count = 0
max_mistake = 6

print("Welcome to Hangman!")


while True: # 게임이 끝날 때까지 반복하는 무한 루프
    print("\nWord: ", " ".join(display_word)) # why \n did not work ????
    print("Guessed words:", guessed_letter) # why no line change ????
    player_input = input ("Guess a letter or the whole world: ").upper() #

    if not player_input.isalpha(): #오직 알파벳만
        print ("Only letters are allowed")
        continue # 입력이 유효하지 않으면 다음 반복으로 넘어

    if len(player_input) ==1:
        if player_input in guessed_letter:
            print("Already guessed that letter")
            continue

        guessed_letter.append(player_input)

        if player_input in secret_word:
           for index, letter in enumerate(secret_word): # enumerate 함수는 글자를 하나씩 꺼내면서 그 글자의 인덱스(index)와 글자(letter)를 함께 가져옴
               if letter == player_input:
                  display_word[index] = player_input
        else:
           mistake_count += 1
           print(f"Wrong! Mistakes : {mistake_count}/{max_mistake}")

    elif player_input == secret_word:
        print(f"You guessed it! The word was {secret_word}")
        break
    else:
        mistake_count += 1
        print(f"Wrong word! Mistakes : {mistake_count}/{max_mistake} ")


    if "_" not in display_word :
      print(f"You won! The word was : {secret_word}")
      break

    if mistake_count == max_mistake:
      print(f"Game over ! The word was : {secret_word}")
      break
```
